```python
import requests
rs = requests.get("https://daodu.tech/archive_article")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(rs.text,"html.parser")
```

```python
urls = soup.find_all('a')
```

```python
urls.pop(0)
```

```python
urls
```

```python
sleep(10)
driver.get(urls[0].get('href'))
sleep(10)
pyautogui.hotkey('ctrl', 'p')
sleep(10)
pyautogui.press('enter')
sleep(10)
pyautogui.press('enter')
```

```python
sleep(5)
for url in urls:
    driver.get(url.get('href'))
    sleep(2)
    pyautogui.hotkey('ctrl', 'p')
    sleep(2)
    pyautogui.press('enter')
    sleep(2)
    pyautogui.press('enter')
```
